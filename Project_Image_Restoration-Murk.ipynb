### Import all libraries

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import cv2
%matplotlib inline 
```

### Restoration1

```python
image=cv2.imread('img3.jfif')
```

```python
imagegray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Input image'), plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(imagegray, cmap = plt.cm.gray)
plt.title('Gray Image'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img1=imagegray.astype('uint8')
bilblur3 = cv2.bilateralFilter(img1, 7, 23, 23)
```

```python
#plt.subplot(121), plt.imshow(imagegray, cmap = plt.cm.gray)
#plt.title('Gray image'), plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(bilblur3, cv2.COLOR_BGR2RGB))
plt.title('Bilateral Blur'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#cv2.imwrite('smoothenedimage3.jpg', bilblur3)
```

```python
mask=cv2.imread('img3-mask-updated.jpg')
cv2.imshow('Mask',mask)
cv2.waitKey(0)
```

```python
maskgray=cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.title('Mask'), plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(maskgray, cv2.COLOR_BGR2RGB))
plt.title('Gray Mask'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
(T, thresh) = cv2.threshold(maskgray, 20, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.subplot(122), plt.imshow(thresh, cmap = plt.cm.gray)
plt.title('Threshold mask'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
restored = cv2.inpaint(bilblur3,thresh,1,cv2.INPAINT_TELEA)
outputres = cv2.inpaint(bilblur3, thresh, 1, cv2.INPAINT_NS)
```

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(restored, cv2.COLOR_BGR2RGB), cmap = plt.cm.gray)
plt.title('Restored Image TELEA'), plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(outputres, cmap = plt.cm.gray)
plt.title('Restored Image NS'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original'), plt.xticks([]),plt.yticks([])
plt.subplot(122), plt.imshow(restored, cmap = plt.cm.gray)
plt.title('Restored TELEA'), plt.xticks([]),plt.yticks([])
#plt.subplot(133), plt.imshow(outputres, cmap = plt.cm.gray)
#plt.title('Restored NS'), plt.xticks([]),plt.yticks([])
plt.show()
```

```python
cv2.imshow("Original vs Restored", np.hstack([bilblur3, restored]))
cv2.waitKey(0)
```
